```python
import cv2 
import glob

import numpy as np

import matplotlib.pyplot as plt

import os

from skimage.io import imread, imshow
import pandas as pd
```

```python
img_dir = '200_Rupees'
data_path = os.path.join(img_dir, '*jpg')
files = glob.glob(data_path)
```

```python


notes_200 = pd.DataFrame([])
for file in files[0:100]:
    img=cv2.imread(file)
    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    edged = cv2.Canny(gray, 10, 250)
    resized = cv2.resize(edged,(100,300))
    note = pd.Series(resized.flatten(),name=file)
    notes_200 = notes_200.append(note)
    notes_200['Target'] = 200
plt.imshow(resized)
```
