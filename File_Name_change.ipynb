```python
import os
 
os.chdir(r'C:\Users\kiran kumar\Documents\Python\Image_processing\Others')
#print(os.getcwd())
COUNT = 1
 
# Function to increment count  
# to make the files sorted.
def increment():
    global COUNT
    COUNT = COUNT + 1
 
 
for f in os.listdir():
    f_name, f_ext = os.path.splitext(f)
    f_name = "Others" + str(COUNT)+'.jpg'
    increment()
 
    new_name = '{}'.format(f_name)
    os.rename(f, new_name) 

```
